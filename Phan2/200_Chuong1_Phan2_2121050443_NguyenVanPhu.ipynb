## THỰC HÀNH 1.5:

Yêu cầu: 
1. Sinh viên đọc ảnh màu images/Thuchanh2.jpg
2. Cho biết chiều cao, chiều rộng và số pixel điểm ảnh.
3. Tách thành các kênh màu B-G-R tương ứng và hiển thị như yêu cầu trong slide
4. Ghi từng kênh màu thành các ảnh riêng vào thư mục images/Saves đặt tên như sau: MSV_blue.jpg; MSV_green.jpg; MSV_red.jpg

```python
#{SINH VIÊN VIẾT CODE VÀO ĐÂY}
#Import các thư viện sử dụng
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
#1. Đọc ảnh với chế độ mặc định: 

img2 = cv2.imread('images/Thuchanh2.jpg')

plt.imshow(img2, cmap='gray')

```

```python
#2. Hiển thị thông tin ảnh:

print(img2.shape)
print(402*728)
print(img2)

# OpenCV lưu trữ ảnh theo thứ tự kênh màu B, G, R nên khi hiển thị lên màn hình ảnh sẽ khác
plt.figure(figsize=(12,6))
plt.subplot(1, 2, 1)
plt.imshow(img2)
plt.title('OpenCV: B-G-R')

#Chuyển về từ hệ màu BGR ---> RGB và hiển thị
img2_rbg = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)

plt.subplot(1, 2, 2)
plt.imshow(img2_rbg)
plt.title('Screen: R-G-B')
plt.show()

```

```python
#3. Tách các kênh màu B, G, R từ ảnh màu

#CÁCH 2: Tách màu BGR
#Lấy kích thước ảnh:
height = img2.shape[0]
width = img2.shape[1]
print('Chiều cao ảnh:',height)
print('Chiều rộng ảnh:',width)
#--------------------------------------------
#Khai báo 2 biến chứa 3 kênh màu G-R-B:
blue=np.zeros((height,width,3),np.uint8)
green=np.zeros((height,width,3),np.uint8)
red=np.zeros((height,width,3),np.uint8)
#-----------------------------------------
#Lấy màu từng pixel tương ứng với mỗi kênh:
for x in range(height):
    for y in range(width):
        B=img2[x,y,0] #Kênh màu Blue: 0
        G=img2[x,y,1] #Kênh màu Green: 1
        R=img2[x,y,2] #Kênh màu Red: 2
        
        blue[x,y,0]=B
        green[x,y,1]=G
        red[x,y,2]=R
```

```python
#Hiển thị ảnh gốc và các kênh màu RGB với Matplotlib

plt.figure(figsize=(15,15))
plt.subplot(2, 2, 1)
plt.imshow(cv2.cvtColor(img2, cv2.COLOR_BGR2RGB))
plt.title('Ảnh RGB')

plt.subplot(2, 2, 2)
plt.imshow(cv2.cvtColor(blue, cv2.COLOR_BGR2RGB))
plt.title('Blue Channel')

plt.subplot(2, 2, 3)
plt.imshow(cv2.cvtColor(green, cv2.COLOR_BGR2RGB))
plt.title('Green Channel')

plt.subplot(2, 2, 4)
plt.imshow(cv2.cvtColor(red, cv2.COLOR_BGR2RGB))
plt.title('Red Channel')
plt.show()


```

```python
#4.Ghi ảnh các kênh màu

cv2.imwrite('images/Saves/2121050443_blue.jpg',blue)
cv2.imwrite('images/Saves/2121050443_green.jpg',green)
cv2.imwrite('images/Saves/2121050443_red.jpg',red)
```

Yêu cầu: Sinh viên tìm hiểu một số hệ màu cơ bản khác:
1. Hệ màu HSV
2. Hệ màu CMYK

1. 
Hệ màu HSV hay không gian màu HSV tạo ra tất cả các màu bằng cách phối hợp các thông tin về màu sắc (hue), độ bão hòa màu (saturation) và ánh sáng (value).

Về không gian màu HSV gồm 3 kênh màu:

- H (Hue): biểu diễn loại màu sắc trong vòng tròn quang phổ với đơn vị đo là góc có giá trị thuộc khoảng [0,179].
- S (Saturation): biểu diễn mức độ bão hòa màu với mức độ thuộc khoảng [0,255].
- V (hay B) (Value hay Bright): biểu diễn độ sáng của màu với dải giá trị thuộc khoảng [0,255].

Các kênh màu thuộc hệ kênh màu HSV có đặc điểm khác với RGB vì có thêm vùng màu và độ bão hòa nên mô tả màu sắc một cách tự nhiên hơn.

<img src='images/HSV.jpg' width='400px'>

Quan sát hình ta có thể thấy được hệ kênh màu HSV như một hình nón khi giá trị của độ sáng bằng 0 thì điểm ảnh là màu đen, còn độ bão hòa màu bằng 0 thì điểm ảnh sẽ màu trắng.



Không gian màu RGB thường được sử dụng trong việc biểu diễn ảnh trên máy tính. Các giá trị màu trên máy tính thường được lưu bằng bộ ba giá trị R, G, B tại mỗi điểm ảnh.

Trong khi đó, không gian màu HSV thường được ứng dụng nhiều vào chỉnh sửa hình ảnh do gần gũi với cách cảm nhận màu sắc của con người. Cụ thể là, khi trao đổi về việc chọn lựa màu sắc, đầu tiên ta sẽ chọn loại màu sắc cơ bản, sau đó thay đổi mức độ bão hòa (hay cường động màu từ nhạt đến đậm). Cuối cùng là điều chỉnh ánh sáng của màu từ tối đến sáng.

2. 
CMYK là từ viết tắt bằng tiếng Anh của cơ chế hệ màu trừ. Nó được sử dụng chủ yếu trong in ấn. Hệ màu CMYK bao gồm các màu sau:

- C = Cyan (xanh lơ).
- M = Magenta (hồng cánh sen).
- Y = Yellow (vàng).
- K = Black (đen).

Sở dĩ dùng ký tự “K” để chỉ màu đen mà không phải “B”, là bởi “B” là ký tự giao ước để chỉ màu xanh (Blue). Ngoài ra, K còn có ý nghĩa là Key - then chốt, chủ yếu.

<img src='images/cmyk.jpg' width='400px'>

Nguyên lý của hệ màu CMYK là hấp thụ ánh sáng. Màu chúng ta nhìn thấy chính là từ phần ánh sáng không hấp thụ. Diễn giải cách khác, chúng hoạt động trên cơ chế vật không tự phát ra ánh sáng. Mà chỉ phản xạ ánh sáng từ các nguồn sáng chiếu tới.

Vậy nên, thay vì thêm độ sáng để có những màu sắc khác nhau. Hệ màu CMYK sẽ loại trừ ánh sáng từ màu trắng (ánh sáng gốc) để tạo ra những màu sắc khác. Mặt khác, 3 màu Cyan, Magenta và Yellow khi kết hợp cũng sẽ tạo ra màu đen.

Màu CMYK thường được sử dụng khi thiết kế phục vụ cho mục đích in ấn các thiết kế như poster, brochure, name card, catalogue, sách hoặc tạp chí,…
