Đọc và hiển thị ảnh đã chọn ở chế độ ảnh màu mặc định, cho biết độ phân giải của ảnh (h x w) và số pixel điểm ảnh

```python
#Import các thư viện sử dụng:
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import cv2 
```

## Yêu cầu 1


```python
#Đọc ảnh 
img = cv2.imread('Pic/Pic_original/Cat_test_9.jpg',1)

#Chiều cao chiều rộng của ảnh
h = img.shape[0]
w = img.shape[1]

#In ra độ phân giải
print('Độ phân giải của ảnh là:', h, 'x', w)

#In ra số pixel điểm ảnh
print('Số pixel điểm ảnh là:', img.size)

#Hiển thị ảnh
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
plt.show()
```

## Yêu cầu 2

```python
#Hiển thị ảnh YUV
plt.subplot(1,2,1)
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2YUV))

#Hiển thị ảnh HSV
plt.subplot(1,2,2)
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2HSV))
plt.show()
```

```python
#Lưu ảnh HSV và YUV vào thư mục Saves
cv2.imwrite('Pic/Saves/2121050443_img_yuv.jpg',cv2.cvtColor(img,cv2.COLOR_BGR2YUV))
cv2.imwrite('Pic/Saves/2121050443_img_hsv.jpg',cv2.cvtColor(img,cv2.COLOR_BGR2HSV))
```

## Yêu cầu 3

```python
#Chiều cao và chiều rộng của ảnh

h = img.shape[0]
w = img.shape[1]

st='Ảnh gốc'

#Thay đổi kích thước ảnh theo tham số nhập vào:
scale = int(input('Nhập vào tỉ lệ % thay đổi kích thước ảnh: '))

#In ra kich thước ảnh gốc
print('1. Kích thước ảnh gốc: ',h ,'x', w)

#Tính toán lại kích thước ảnh:
h_new = int((h * scale)/100)
w_new = int((w * scale)/100)

#In ra kích thước ảnh sau khi thay đổi
print('2. Kích thước ảnh sau thay đổi: ',h_new ,'x', w_new)

st_new = 'Ảnh Đã Scale'

d_sv = (w_new,h_new)

#Nếu giá trị <100 hiển thị ảnh thu nhỏ bằng thuật toán INTER_AREA
if scale < 100:
    img_new = cv2.resize(img,
                        d_sv,
                        interpolation=cv2.INTER_AREA)
    print('--> NOTE: Bạn sẽ thu nhỏ ảnh(Thuật toán: cv2.INTER_AREA)')
    
#Nếu giá trị >=100 hiển thị ảnh phóng to bằng thuật toán INTER_CUBIC
else: 
    img_new = cv2.resize(img,
                        d_sv,
                        interpolation=cv2.INTER_CUBIC)
    print("--> NOTE: Bạn sẽ phóng to ảnh(Thuật toán: cv2.INTER_CUBIC)")

#Ảnh gốc ban đầu
plt.figure(figsize=(15,7))
plt.subplot(1,2,1)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title(st)

#Ảnh sau khi thay đổi kích thước
plt.subplot(1,2,2)
plt.imshow(cv2.cvtColor(img_new, cv2.COLOR_BGR2RGB))
plt.title(st_new)

plt.show()
```

```python
#Lưu ảnh ra thư mục Saves
cv2.imwrite('Pic/Saves/2121050443_img_scale.jpg',img_new)
```
